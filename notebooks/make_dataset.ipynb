```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../data/processed/Dronninglund_treated_data_and_flow_rates.csv', parse_dates=True, index_col=0) 
df
```

```python
water_sensors = ['T_{:04.1f}'.format(height) for height in np.linspace(16, 0.5, 32)]
df_water_sensors = df[water_sensors]
df_water_sensors.columns = [int(float(c[2:])-0.5) for c in water_sensors]
df_water_sensors = df_water_sensors.groupby(df_water_sensors.columns, axis='columns').mean()
# df_water_sensors = (df_water_sensors - T_MIN) / (T_MAX - T_MIN)
df_water_sensors
```

```python
inputs = pd.concat([ df.F_top_vol * df.FR_top_c, df.F_bot_vol * df.FR_bot_c, df.F_mid_vol * df.FR_mid_c, df.T_top, df.T_mid, df.T_bot, df.F_top_vol, df.F_bot_vol,df.F_mid_vol], axis=1)
inputs.columns = ['Ftop', 'Fbot', 'Fmid', 'Ttop', 'Tmid', 'Tbot', 'FtopVol', 'FbotVol', 'FmidVol']
# inputs[['T_top', 'T_mid', 'T_bot']] = (inputs[['T_top', 'T_mid', 'T_bot']] - T_MIN) / (T_MAX - T_MIN)
# inputs[['F_top', 'F_bot', 'F_mid']] = inputs[['F_top', 'F_bot', 'F_mid']] / inputs[['F_top', 'F_bot', 'F_mid']].max()
inputs
```

```python
variables = df[['temp_dry', 'Tg_25', 'Tg_20', 'Tg_15', 'Tg_10']]
```

```python
start_date = '2014-05-01' # Nothing realy happens before ...
freq = '1h'
inputs = inputs[start_date:].resample(freq).mean()
variables = variables[start_date:].resample(freq).mean()
df_water_sensors = df_water_sensors[start_date:].resample(freq).mean()
t = (df_water_sensors.index - df_water_sensors.index[0]).total_seconds() / 3600
df_water_sensors['t'] = t


inputs['FtopIn'] = inputs.Ftop * (inputs.Ftop > 0)
inputs['FmidIn'] = inputs.Fmid * (inputs.Fmid > 0)
inputs['FbotIn'] = inputs.Fbot * (inputs.Fbot > 0)

inputs['FtopOut'] = -inputs.Ftop * (inputs.Ftop < 0)
inputs['FmidOut'] = -inputs.Fmid * (inputs.Fmid < 0)
inputs['FbotOut'] = -inputs.Fbot * (inputs.Fbot < 0)

inputs.to_csv('../data/processed/dronninglund/inputs.csv')
df_water_sensors.to_csv('../data/processed/dronninglund/water_sensors.csv')
variables.to_csv('../data/processed/dronninglund/variables.csv')



```
