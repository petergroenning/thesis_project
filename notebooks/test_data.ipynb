```python
N = 14 * 2
M = len(df)

tempA = np.zeros((N, M))
tempB = np.zeros((N, M))
cols = df.columns
for i in range(N // 2):
    tempA[i*2] = df[cols[i]].values
    tempA[i*2+1]   = df[cols[i + N // 2]].values 

    tempB[i*2] = df[cols[i + N]].values
    tempB[i*2+1]   = df[cols[i + N + N // 2]].values

tempA = (tempA - tempA.min()) / (tempA.max() - tempA.min())
tempB = (tempB - tempB.min()) / (tempB.max() - tempB.min())
```

```python
fig, ax = plt.subplots(3, 1, figsize=(20, 10))
f = ax[0].imshow(tempA, aspect='auto', interpolation='none', cmap='Reds')
cbar = plt.colorbar(f)
ax[0].plot(df.U_OUT.values)
ax[0].plot(df.U_IN.values)

f = ax[1].imshow(tempB, aspect='auto', interpolation='none', cmap='Reds')
cbar = plt.colorbar(f)
ax[1].plot(df.U_OUT.values)
ax[1].plot(df.U_IN.values)


D = tempA - tempB
f = ax[2].imshow(D, aspect='auto', interpolation='none', cmap='viridis')
cbar = plt.colorbar(f)
ax[2].plot(df.U_OUT.values)
ax[2].plot(df.U_IN.values)


fig.tight_layout()
```

# Euler 


```python
# euler simulation
import numpy as np
import matplotlib.pyplot as plt

f1 = lambda x,y: -0.3*x - 0.1
f2 = lambda x,y: 0.3*x

x = 1
y = 0
dt = 0.1
T = 100
N = int(T/dt)
X = np.zeros(N)
Y = np.zeros(N)
for i in range(N):
    x += f1(x,y)*dt
    y += f2(x,y)*dt
    X[i] = x
    Y[i] = y

fig, ax = plt.subplots(1, 1)
ax.plot(X, label = 'x')
ax.plot(Y, label = 'y')
ax.legend()

```
